```python
import pandas as pd
import numpy as np

import datetime
```

```python
h_trs = pd.read_csv("../input/historical_transactions.csv", usecols=["card_id", "purchase_date"], parse_dates=["purchase_date"])
n_trs = pd.read_csv("../input/new_merchant_transactions.csv", usecols=["card_id", "purchase_date"], parse_dates=["purchase_date"])
```

```python
all_trs = pd.concat([h_trs, n_trs], axis=0).reset_index(drop=True)
all_trs.purchase_date = all_trs.purchase_date.apply(lambda x:x.date())
```

```python
all_trs.head()
```

second_active

```python
all_trs["one"] = 1
```

```python
df_purchase_pattern = all_trs.groupby("card_id")["purchase_date"].min().reset_index()
df_purchase_pattern.columns = ["card_id", "second_active_date"]

df_purchase_pattern["elapsed_second_absolute"] = (df_purchase_pattern.second_active_date - datetime.date(2010, 1, 1)).dt.days
```

```python
df_purchase_pattern.head()
```

```python
df_all_purchase = pd.pivot_table(all_trs, columns=["purchase_date"], index=["card_id"], values=["one"], aggfunc=sum, fill_value=0)
df_all_purchase.columns = [col[1] for col in df_all_purchase.columns.values]
df_all_purchase = df_all_purchase.reset_index()
```

```python
df_all_purchase.head()
```

```python
third_active_date = [datetime.date(2010, 1, 1) for i in range(df_all_purchase.shape[0])]

for idx, r in df_all_purchase.iterrows():
    all_purcase_date_idx = np.where(r[1:]>0)[0] + 1
    if len(all_purcase_date_idx) >= 2:
        third_idx = all_purcase_date_idx[1]
    else:
        third_idx = all_purcase_date_idx[0]
    third_active_date[idx] = df_all_purchase.columns[third_idx]
```

```python
df_third_active = pd.DataFrame({"card_id":df_all_purchase.card_id, "third_active_date":third_active_date})
```

```python
df_purchase_pattern = pd.merge(df_purchase_pattern, df_third_active, on="card_id", how="outer")
```

```python
df_purchase_pattern.head()
```

```python
df_purchase_pattern["elapsed_third_absolute"] = (df_purchase_pattern.third_active_date - datetime.date(2010, 1, 1)).dt.days
```

```python
df_purchase_pattern["diff_second_and_third"] = (df_purchase_pattern.third_active_date - df_purchase_pattern.second_active_date).dt.days
```

```python
purchase_freq = [0 for i in range(df_all_purchase.shape[0])]

for idx, r in df_all_purchase.iterrows():
    all_purcase_date_idx = np.where(r[1:]>0)[0] + 1
    if len(all_purcase_date_idx) >= 2:
        purchase_count = len(all_purcase_date_idx)
        
        top_idx = all_purcase_date_idx[0]
        end_idx = all_purcase_date_idx[-1]
        
        top_date = df_all_purchase.columns[top_idx]
        end_date = df_all_purchase.columns[end_idx]
        
        purchase_term = (end_date - top_date).days
        
        freq = 1.0 * purchase_term / purchase_count
    else:
        freq = 0
    purchase_freq[idx] = freq
```

```python
df_purchase_freq = pd.DataFrame({"card_id":df_all_purchase.card_id, "purchase_freq":purchase_freq})
```

```python
df_purchase_pattern = pd.merge(df_purchase_pattern, df_purchase_freq, on="card_id", how="outer")
```

```python
list_purchase_term = [0 for i in range(df_all_purchase.shape[0])]

for idx, r in df_all_purchase.iterrows():
    all_purcase_date_idx = np.where(r[1:]>0)[0] + 1
        
    top_idx = all_purcase_date_idx[0]
    end_idx = all_purcase_date_idx[-1]
        
    top_date = df_all_purchase.columns[top_idx]
    end_date = df_all_purchase.columns[end_idx]
        
    purchase_term = (end_date - top_date).days
    
    list_purchase_term[idx] = purchase_term
```

```python
df_purchase_term = pd.DataFrame({"card_id":df_all_purchase.card_id, "purchase_term":list_purchase_term})
```

```python
df_purchase_pattern = pd.merge(df_purchase_pattern, df_purchase_term, on="card_id", how="outer")
```

```python
h_trs = pd.read_csv("../input/historical_transactions.csv", usecols=["card_id", "purchase_amount", "purchase_date"], parse_dates=["purchase_date"])
n_trs = pd.read_csv("../input/new_merchant_transactions.csv", usecols=["card_id", "purchase_amount", "purchase_date"], parse_dates=["purchase_date"])
```

```python
all_trs = pd.concat([h_trs, n_trs], axis=0).reset_index(drop=True)
all_trs.purchase_date = all_trs.purchase_date.apply(lambda x:x.date())
```

```python
np.min(all_trs.purchase_amount)
```

```python
all_trs.purchase_amount = np.log(all_trs.purchase_amount + 100)
```

```python
df_all_purchase_amount = pd.pivot_table(all_trs, columns=["purchase_date"], index=["card_id"], values=["purchase_amount"], aggfunc=sum, fill_value=0)
df_all_purchase_amount.columns = [col[1] for col in df_all_purchase_amount.columns.values]
df_all_purchase_amount = df_all_purchase_amount.reset_index()
```

```python
df_all_purchase_amount
```

```python
from sklearn.decomposition import NMF
sparse_matrix = df_all_purchase_amount[df_all_purchase_amount.columns[1:]].values

N_COMP =10
model = NMF(n_components=N_COMP, init='random', random_state=0) 
embedded = model.fit_transform(sparse_matrix)

df = pd.DataFrame(embedded, columns=["NMF_comp_purchase_amount_{}".format(i) for i in range(1,N_COMP+1)])
df["card_id"] = df_all_purchase_amount.card_id.values
```
