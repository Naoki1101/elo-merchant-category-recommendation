```python
import numpy as np
import pandas as pd
```

```python
h_trs = pd.read_csv("../input/historical_transactions.csv", usecols=["card_id"])
n_trs = pd.read_csv("../input/new_merchant_transactions.csv", usecols=["card_id"])
```

```python
df_h_trs_index = h_trs.reset_index().groupby("card_id")["index"].min().reset_index()
df_h_trs_index.columns = ["card_id", "h_trs_index"]
```

```python
df_n_trs_index = n_trs.reset_index().groupby("card_id")["index"].min().reset_index()
df_n_trs_index.columns = ["card_id", "n_trs_index"]
```

```python
df_trs_index = pd.merge(df_h_trs_index, df_n_trs_index, on="card_id", how="left")
```

```python
df_trs_index = df_trs_index.sort_values(by="h_trs_index").reset_index(drop=True)
df_trs_index
```

```python
df_trs_index.to_csv("../input/trs_index_min.csv", index=False)
```
