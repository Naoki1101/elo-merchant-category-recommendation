```python
import pandas as pd
import numpy as np

from tqdm import tqdm

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_train = pd.read_csv("./input/train.csv")
```

```python
merchants = pd.read_csv("./input/merchants.csv")
```

```python
h_trs = pd.read_csv("./input/historical_transactions.csv")
```

```python
n_trs = pd.read_csv("./input/new_merchant_transactions.csv")
```

```python
df_sales_feature = merchants[["merchant_id", "avg_sales_lag3", "avg_purchases_lag3"]]
df_sales_feature.avg_sales_lag3 = df_sales_feature.avg_sales_lag3.apply(lambda x: np.log(x+100))
df_sales_feature["unit_price"] = df_sales_feature.avg_sales_lag3 / df_sales_feature.avg_purchases_lag3
```

```python
merged_h_trs = pd.merge(
    h_trs[["card_id", "merchant_id", "purchase_amount"]],
    df_sales_feature,
    on="merchant_id",
    how="left"
)

merged_n_trs = pd.merge(
    n_trs[["card_id", "merchant_id", "purchase_amount"]],
    df_sales_feature,
    on="merchant_id",
    how="left"
)
```

```python
merged_h_trs.purchase_amount = np.log(merged_h_trs.purchase_amount.values + 100)
merged_h_trs["purchase_level"] = merged_h_trs.purchase_amount / merged_h_trs.avg_sales_lag3
merged_h_trs["estimated_items"] = merged_h_trs.purchase_amount / merged_h_trs.unit_price

merged_n_trs.purchase_amount = np.log(merged_n_trs.purchase_amount.values + 100)
merged_n_trs["purchase_level"] = merged_n_trs.purchase_amount / merged_n_trs.avg_sales_lag3
merged_n_trs["estimated_items"] = merged_n_trs.purchase_amount / merged_n_trs.unit_price
```

```python
agg_func = {
    'avg_sales_lag3': ['std', 'max', 'min', 'mean'],
    'avg_purchases_lag3': ['std', 'max', 'min', 'mean'],
    'unit_price': ['std', 'max', 'min', 'mean'],
    'purchase_level': ['std', 'max', 'min', 'mean']
}

merchant_sales_summary = merged_h_trs.groupby(["card_id"]).agg(agg_func)
new_merchant_sales_summary = merged_n_trs.groupby(["card_id"]).agg(agg_func)
```

```python
merchant_sales_summary.columns = ["_".join(col).strip() for col in merchant_sales_summary.columns.values]
new_merchant_sales_summary.columns =  ["new" + "_".join(col).strip() for col in new_merchant_sales_summary.columns.values]
```

```python
merchant_sales_summary = merchant_sales_summary.reset_index()
new_merchant_sales_summary = new_merchant_sales_summary.reset_index()
```

```python
all_merchant_sales_summary = pd.merge(
    merchant_sales_summary,
    new_merchant_sales_summary, 
    on="card_id", 
    how="outer"
)
```
