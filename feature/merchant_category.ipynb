```python
import os
from collections import defaultdict
from csv import DictReader
import math

import numpy as np
import pandas as pd
```

```python
usecols = [
    "card_id",
    "merchant_id",
    "category_1", 
    "category_2",
    "category_3"
]
```

```python
h_trs = pd.read_csv("../input/historical_transactions.csv", usecols=usecols)
n_trs = pd.read_csv("../input/new_merchant_transactions.csv", usecols=usecols)
merchants = pd.read_csv("../input/merchants.csv")
```

```python
trs = pd.concat([h_trs, n_trs], axis=0).reset_index(drop=True)
```

```python
trs_category = pd.merge(
    trs, 
    merchants[["merchant_id", "category_4"]], 
    on="merchant_id", 
    how="left"
)[["card_id", "category_1",  "category_2", "category_3", "category_4"]]
```

```python
trs_category_dummy = pd.get_dummies(data=trs_category, columns=["category_1", "category_2", "category_3", "category_4"])
trs_category_dummy
```

```python
trs_category_dummy.columns
```

```python
agg_func = {
    'category_1_N': ["sum", "mean", "std"], 
    'category_1_Y': ["sum", "mean", "std"], 
    'category_2_1.0': ["sum", "mean", "std"],
    'category_2_2.0': ["sum", "mean", "std"], 
    'category_2_3.0': ["sum", "mean", "std"], 
    'category_2_4.0': ["sum", "mean", "std"], 
    'category_2_5.0': ["sum", "mean", "std"],
    'category_3_A': ["sum", "mean", "std"], 
    'category_3_B': ["sum", "mean", "std"], 
    'category_3_C': ["sum", "mean", "std"], 
    'category_4_N': ["sum", "mean", "std"],
    'category_4_Y': ["sum", "mean", "std"]
}

trs_category_summary = trs_category_dummy.groupby("card_id").agg(agg_func)
```

```python
trs_category_summary.columns = ["_".join(i).strip() for i in trs_category_summary.columns.values]
```

```python
trs_category_summary = trs_category_summary.reset_index()
```

```python
trs_category_summary.to_csv("../input/trs_category_summary.csv", index=False)
```
