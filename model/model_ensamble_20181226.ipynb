```python
import numpy as np
import pandas as pd
import os
```

```python
os.listdir("../output/")
```

```python
def ensamble():
    df = pd.read_csv("../output/first_submission.csv", usecols=["card_id"])
    list_output = os.listdir("../output/")
    counter = 1
    for file in list_output:
        if "submission_model2018" in file:
            data = pd.read_csv("../output/{}".format(file))
            data.columns = ["card_id", file.split(".csv")[0]]
            df = pd.concat([df, data[[file.split(".csv")[0]]]], axis=1)
        counter += 1
    return df
```

```python
df = ensamble()
```

```python
df
```

```python
df["target"] = np.mean(df[df.columns[1:]], axis=1)
```

```python
submission = df[["card_id", "target"]]
```

```python
submission.to_csv("../output/ensamble_20181226.csv", index=False)
```
