```python
import numpy as np
import pandas as pd
```

```python
df_submission = pd.read_csv("../output/submission_model20181216.csv")
```

```python
df_outliers_prob = pd.read_csv("../output/outliers_prob_20190114.csv")
```

```python
df = pd.merge(df_submission, df_outliers_prob, on="card_id", how="inner")
df.sort_values(by="outliers_prob", ascending=False)
```

```python
outliers_idx = df[(df.target <= -15)&(df.outliers_prob >= 0.80)].index
outliers_idx
```

```python
df["target"].iloc[outliers_idx] = -20
```

```python
df.describe()
```

```python
df[["card_id", "target"]].to_csv("../output/submission_pinpoint_model_20190115.csv", index=False)
```
